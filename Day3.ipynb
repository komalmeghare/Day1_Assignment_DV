### Question 1: Plot using sns.relplot, line plot, box plot on fmri data set

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
df = sns.load_dataset("fmri")
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['region'].value_counts()
```

```python
sns.relplot(x='timepoint',y='signal',data=df,color='r')
```

```python
sns.relplot(x='timepoint',y='signal',data=df,hue='event',height=10)
```

```python
sns.relplot(x='timepoint',y='signal',data=df,hue='event',style='region',height=10)
```

```python
sns.relplot(x='subject',y='signal',data=df,color='g')
```

```python
sns.relplot(x='timepoint',y='signal',data=df,hue='region',kind='line',style='event')
```

```python
sns.relplot(x='subject',y='signal',data=df,color='purple',kind='line',style='event')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', style ='region', kind = 'line', color='r', data = df)
```

```python
sns.relplot(x='timepoint',y='signal',data=df,color='b',kind='line',hue='event',style ='region')
```

```python
sns.catplot(x='event',y='signal',kind='box',data=df,hue='region')
```

```python
sns.boxplot(x='region',y='signal',data=df,hue='event')
```
