```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sine waveform", color='r',fontsize=15)
plt.xlabel("X-axis", color='r',fontsize=13)
plt.ylabel("sin(x)", color='r',fontsize=13)
plt.plot(x,y,color='b')
```

```python
x = np.arange(0,10)
y = x*x
plt.plot(x,y,'ro--',linewidth=1,markersize=10)
plt.title("Squaring X value",color='b',fontsize=15)
plt.xlabel('X-axis',color='b',fontsize=13)
plt.ylabel('Square of X',color='b',fontsize=13)
```

```python
x=[1,2,3,4,5,6,7]
y=[160,150,140,145,175,165,180]
plt.title("SALES-TREND OVER DAYS",color='r',fontsize=17)
plt.xlabel("Days of the week",color='r',fontsize=15)
plt.ylabel("Sales of Company",color='r',fontsize=15)
plt.plot(x,y,'g*--',linewidth=1,markersize=10)

#plt.tick_params(axis='both',lablesize=10)
```
